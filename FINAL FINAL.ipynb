```python
selected_columns = ['passer_id', 'passer', 'season', 'qb_epa_last',
       'epa_mean', 'ypa_last', 'inteception_last', 'pass_td_last']
```

```python
subset_qb_data = qb_data[selected_columns]
```

```python
import pandas as pd

# Assuming qb_data is your DataFrame
# Replace 'season' with the actual column name in your DataFrame for the season information

# Define weights for each metric for each year (normalized to sum up to 1)
yearly_weights = {
    2017: {'epa_mean': 0.2, 'inteception_last': 0.2, 'qb_epa_last': 0.2, 'pass_td_last': 0.2, 'ypa_last': 0.2},
    2018: {'epa_mean': 0.2, 'inteception_last': 0.2, 'qb_epa_last': 0.2, 'pass_td_last': 0.2, 'ypa_last': 0.2},
    2019: {'epa_mean': 0.2, 'inteception_last': 0.2, 'qb_epa_last': 0.2, 'pass_td_last': 0.2, 'ypa_last': 0.2},
    2020: {'epa_mean': 0.2, 'inteception_last': 0.2, 'qb_epa_last': 0.2, 'pass_td_last': 0.2, 'ypa_last': 0.2},
    2021: {'epa_mean': 0.2, 'inteception_last': 0.2, 'qb_epa_last': 0.2, 'pass_td_last': 0.2, 'ypa_last': 0.2},
    2022: {'epa_mean': 0.2, 'inteception_last': 0.2, 'qb_epa_last': 0.2, 'pass_td_last': 0.2, 'ypa_last': 0.2},
    2023: {'epa_mean': 0.2, 'inteception_last': 0.2, 'qb_epa_last': 0.2, 'pass_td_last': 0.2, 'ypa_last': 0.2},
}

# Apply weights based on the season
subset_qb_data['composite_score'] = (
    subset_qb_data['epa_mean'] * subset_qb_data['season'].map(yearly_weights).apply(lambda x: x['epa_mean']) -
    subset_qb_data['inteception_last'] * subset_qb_data['season'].map(yearly_weights).apply(lambda x: x['inteception_last']) +
    subset_qb_data['qb_epa_last'] * subset_qb_data['season'].map(yearly_weights).apply(lambda x: x['qb_epa_last']) +
    subset_qb_data['pass_td_last'] * subset_qb_data['season'].map(yearly_weights).apply(lambda x: x['pass_td_last']) +
    subset_qb_data['ypa_last'] * subset_qb_data['season'].map(yearly_weights).apply(lambda x: x['ypa_last'])
)

# Display the modified DataFrame
print(subset_qb_data[['passer', 'season', 'composite_score']])
```

```python
# Sort DataFrame by 'composite_score' in descending order
top_qbs = subset_qb_data.sort_values(by='composite_score', ascending=False)

# Select the top 10 quarterbacks
top_10_qbs = top_qbs.head(10)

# Display the top 10 quarterbacks
print(top_10_qbs[['season','passer', 'composite_score']])
```

```python

# Sort DataFrame by 'composite_score' in descending order
top_qbs = subset_qb_data.sort_values(by=['season', 'composite_score'], ascending=[True, False])

# Select the top quarterback for each season
top_qbs_by_season = top_qbs.groupby('season').head(1)

# Display the top quarterback for each season
print(top_qbs_by_season[['season', 'passer', 'composite_score']])
```

```python

import matplotlib.pyplot as plt

# Assuming qb_data is your DataFrame
# (make sure to replace 'season' with the actual column name in your DataFrame for the season information)

# Filter data for the 2017 season
top_10_2017 = subset_qb_data[subset_qb_data['season'] == 2017].sort_values(by='composite_score', ascending=False).head(10)

# Plotting the composite scores for the top 10 players in 2017
plt.figure(figsize=(12, 6))
plt.bar(top_10_2017['passer'], top_10_2017['composite_score'], color='blue')
plt.xlabel('Quarterback')
plt.ylabel('Composite Score')
plt.title('Top 10 Composite Scores for Quarterbacks in 2017')

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45, ha='right')

# Adjust layout for better appearance
plt.tight_layout()

# Show the plot
plt.show()

```
